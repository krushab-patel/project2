```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = pd.read_excel("Superstore_USA.xlsx")
```

```python
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data['Product Base Margin'].fillna(data['Product Base Margin'].mean(),inplace = True)
```

# order priority

```python
data['Order Priority'].value_counts()
```

```python
data['Order Priority'].unique()
```

```python
data['Order Priority'] = data['Order Priority'].replace("Critical ","Critical")
```

```python
data['Order Priority'].value_counts()
```

```python
plt.figure(figsize=(5,3))
df = sns.countplot(data = data, x = 'Order Priority') 
plt.title("counts of order priority")
plt.savefig("count of order priority.jpg")
```

# ship mode

```python
data['Ship Mode'].value_counts()
```

```python
x = data['Ship Mode'].value_counts().index
y = data['Ship Mode'].value_counts().values
```

```python
plt.figure(figsize=(5,4))
plt.pie(y,labels = x,startangle = 60,autopct="%0.2f%%")
plt.legend(loc=2)
plt.show()
```

```python
plt.figure(figsize=(5,4))
df = sns.countplot(data = data,x = 'Ship Mode',hue= 'Product Category')
```
